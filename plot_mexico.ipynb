### Load libraries.

```python
import geopandas as geopd
import pandas as pd
```

### Load Mexico shape file. Source: Marco Geoestadístico INEGI (2023).
#### URL: https://www.inegi.org.mx/temas/mg/ 

```python
# INEGI data
gdf = geopd.read_file("../dataset/Mexico/geometry/2023_1_00_MUN/2023_1_00_MUN.shp")
```

### Display data frame entries

- CVEGEO: Geoestatistical code provided by INEGI.
- CVE_ENT: State code provided by INEGI.
- CVE_MUN: Municipality code provided by INEGI.
- NOMGEO: Municipality name provided by INEGI.
- geometry: Municipality boundary provided by INEGI (polygons).

```python
display(gdf)
```

### Mexico has 32 states. Mexico City is the capital of the country.

```python
# Plot Mexico's map labeling each state by a color. 
# States are identified bya a CVE_ENT code

ax = gdf.plot(figsize = (60, 60),column="CVE_ENT",cmap="Blues")
ax.set_axis_off()
```

### Each Mexican state is subdivided into municipalities.

```python
# Plot Mexico's map labeling each municipality by a color. 
# Municipalities  are identified bya a CVE_MUN code.
ax =gdf.plot(figsize = (20, 20),column="CVE_MUN",cmap="Blues")

# Include municipalities outlines.
gdf.boundary.plot(ax=ax,figsize = (20, 20), edgecolor='midnightblue')
ax.set_axis_off()
```
