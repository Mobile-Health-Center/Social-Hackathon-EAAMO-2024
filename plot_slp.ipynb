### Load libraries.

```python
import geopandas as geopd
import pandas as pd
```

### Load San Luis Potosi state shape file. Source: Marco Geoestadístico INEGI (2023).
#### URL: https://www.inegi.org.mx/temas/mg/ 

```python
# INEGI data
gdf = geopd.read_file("../dataset/San Luis Potosi/geometry/2023_1_24_MUN/2023_1_24_MUN.shp")
```

### Display data frame entries

- CVE_ENT: State code provided by INEGI.
- CVE_MUN: Municipality code provided by INEGI.
- NOMGEO: Municipality name provided by INEGI.
- geometry: Municipality boundary provided by INEGI (polygons).

```python
display(gdf)
```

### San Luis Potosi has 58 municipalities. San Luis Potopsí is the capital of the country.

```python
# Plot Mexico's map labeling each state by a color. 
# Municipalities are identified bya a CVE_MUN code

ax = gdf.plot(figsize = (60, 60),column="CVE_MUN", cmap = "Blues")
ax.set_axis_off()

# Include municipalities outlines.
gdf.boundary.plot(ax=ax,figsize = (20, 20), edgecolor='midnightblue')
ax.set_axis_off()
```
